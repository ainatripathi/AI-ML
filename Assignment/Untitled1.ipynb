```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.DataFrame({'hours':[2,3,4,5,6],
                 'attendance':[60,70,15,80,90],
                 'result':['fail','fail','pass','pass','pass']})
```

```python
x1=df[['hours','attendance']]
y1=df['result']
```

```python
rf1=RandomForestClassifier(n_estimators=100,random_state=42)
rf1.fit(x1,y1)
```

```python
pred=rf1.predict([[4,85]])
print("Prediction: ",pred)
```

```python
plt.scatter(df['hours'],df['attendance'])
plt.xlabel("Hours")
plt.ylabel("Attendance")
```

```python
data2 = pd.DataFrame({
    'Income': [30000, 40000, 50000, 60000, 70000],
    'Credit_Score': [600, 650, 700, 750, 800],
    'Loan_Status': ['No', 'No', 'Yes', 'Yes', 'Yes']
})

X2 = data2[['Income', 'Credit_Score']]
y2 = data2['Loan_Status']

rf2 = RandomForestClassifier(n_estimators=100, random_state=42)
rf2.fit(X2, y2)

prediction2 = rf2.predict([[55000, 720]])
print("Question 2 Prediction:", prediction2[0])
plt.scatter(data2['Income'],data2['Credit_Score'])
plt.xlabel("Income")
plt.ylabel("Credit_Score")
```

```python
data3 = pd.DataFrame({
    'Experience': [1, 2, 3, 4, 5, 6, 7],
    'Performance_Score': [50, 55, 60, 65, 75, 85, 90],
    'Promotion': ['No', 'No', 'No', 'No', 'Yes', 'Yes', 'Yes']
})

X3 = data3[['Experience', 'Performance_Score']]
y3 = data3['Promotion']

rf3 = RandomForestClassifier(n_estimators=100, random_state=42)
rf3.fit(X3, y3)

prediction3 = rf3.predict([[8, 88]])
print("Question 3 Prediction:", prediction3[0])
plt.scatter(data3['Experience'],data3['Performance_Score'])
plt.xlabel("Experience")
plt.ylabel("Performance_Score")
```
