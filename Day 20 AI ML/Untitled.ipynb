```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('placement.csv')
df
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package (in LPA)')
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
x
y
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2) #0.2->test size 20%
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train) #method fit is used to train the model
#method is trained with small data so gets trained fast
lr.predict(x_test.values) #to predict the test values of x
```

```python
lr.predict(x_test.iloc[1].values.reshape(1,1))
x_test
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package (in LPA)')
```
