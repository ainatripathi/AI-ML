```python
from sklearn.datasets import make_classification
import numpy as np
x,y=make_classification(n_samples=100,n_features=2, n_informative=1, n_redundant=0, n_classes=2,n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)
```

```python
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
np.insert(x,0,1,axis=1)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
```

```python
def step(z):
    return 1 if z >= 0 else 0
```

```python
def perceptron(X,y):

    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1]) #coefficient
    lr = 0.1

    for i in range(1000): #epoch
        j = np.random.randint(0,100) #random values
        y_hat = step(np.dot(X[j],weights)) #dot product >0
        weights = weights + lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:]
```

```python
intercept_,coef_ = perceptron(x,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1]) #m=-a/b
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(x,y)
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input + b
```

```python
# Re-execute perceptron to ensure correct final intercept_ and coef_ are used
intercept_,coef_ = perceptron(x,y)

# Recalculate perceptron line parameters to ensure correctness
m_perceptron = -(coef_[0]/coef_[1])
b_perceptron = -(intercept_/coef_[1])
y_input_perceptron = m_perceptron*x_input + b_perceptron

# Recalculate logistic regression line parameters to ensure correctness
m_logistic = -(lor.coef_[0][0]/lor.coef_[0][1])
b_logistic = -(lor.intercept_[0]/lor.coef_[0][1]) # lor.intercept_ is an array, need to access the scalar
y_input_logistic = m_logistic*x_input + b_logistic

plt.figure(figsize=(10,6))
plt.plot(x_input, y_input_perceptron, color='red', linewidth=3, label='Perceptron Decision Boundary')
plt.plot(x_input, y_input_logistic, color='black', linewidth=3, label='Logistic Regression Decision Boundary')
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundaries of Perceptron and Logistic Regression')
plt.legend()
# Removing plt.ylim to view the full extent of the lines, or set a wider range
# plt.ylim(-200, 200) # Example of a wider range if you want to keep limits
plt.grid(True)
plt.show()
```
