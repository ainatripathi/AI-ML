```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
```

```python
#Line Graph
price=[48000,54000,49000,47000,54000,42000]
year=[2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
batsman=pd.read_csv('sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit vs Virat Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='purple')
plt.plot(batsman['index'],batsman['RG Sharma'],color='cyan')
plt.title('Rohit vs Virat Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='purple',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='cyan',linestyle='dashdot')
plt.title('Rohit vs Virat Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='purple',linestyle='dotted',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='cyan',linestyle='dashdot',linewidth=1.5)
plt.title('Rohit vs Virat Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='purple',linestyle='dotted',linewidth=3,marker='o',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='cyan',linestyle='dashdot',linewidth=1.5,marker='o',markersize=8)
plt.title('Rohit vs Virat Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='purple',linestyle='dotted',linewidth=3,marker='o',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='cyan',linestyle='dashdot',linewidth=1.5,marker='o',markersize=8,label='Rohit')
plt.title('Rohit vs Virat Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.legend(loc='upper left')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='purple',linestyle='dotted',linewidth=3,marker='o',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='cyan',linestyle='dashdot',linewidth=1.5,marker='o',markersize=8,label='Rohit')
plt.title('Rohit vs Virat Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.legend(loc='upper left')
plt.grid()
plt.show() #to show graph on website; used in frontend
```
