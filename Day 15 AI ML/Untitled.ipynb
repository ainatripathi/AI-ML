```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size',kind='scatter')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size',kind='line')
```

```python
gap=px.data.gapminder()
gap
temp_df=gap[gap['country']=='Japan']
temp_df
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
temp_df=gap[gap['country'].isin(['India','Brazil','China'])] 
temp_df
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',style='continent',size='country',hue='country')
```
